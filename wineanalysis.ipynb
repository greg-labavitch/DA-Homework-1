```python
#Dependencies

import matplotlib.pyplot as plt 
import pandas as pd 
import scipy.stats as st
import numpy as np
import seaborn as sns
```

```python
# datafiles for red wine
# this is on mac
red_wine_read = "Resources/winequality-red.csv"

red_wine=pd.read_csv(red_wine_read, delimiter=";")

red_wine_df = pd.DataFrame(red_wine)
red_wine_df.head()
red_wine_df.count()
```

```python
# datafiles for white wine
# this is on mac
white_wine_read= "Resources/winequality-white.csv"

white_wine = pd.read_csv(white_wine_read, delimiter=";")
white_wine_df= pd.DataFrame(white_wine)
white_wine_df.head()
white_wine_df.count()
```

```python
red_white_wine = pd.merge(red_wine_df, white_wine_df, how="outer", indicator = True)
red_white_wine

```

```python
red_white_wine2 = red_white_wine.drop(red_white_wine[red_white_wine["_merge"]=='both'].index)
red_white_wine2
```

```python
#Check info of merged data frame
red_white_wine2.info()
```

```python
red_white_wine2.describe()
```

```python
# mean of overall merged wine quality (red_white)

overall_wine_quality = np.mean(red_white_wine2["quality"])
overall_wine_quality
```

```python
wine_averages = red_white_wine2.groupby("_merge")["quality"].mean()
wine_averages_df = pd.DataFrame(wine_averages)
wine_averages_df
```

```python
red_quality = wine_averages_df.loc["left_only" , "quality"]
white_quality = wine_averages_df.loc["right_only" , "quality"]

print(red_quality)
print(white_quality)
```

```python
bar_labels = ["Red Quality","White Quality","Overall Quality"]
qualities = [red_quality, white_quality, overall_wine_quality]

```

```python
# bar chart showing overall quality
# bar chart showing red wine quality
# bar chart showing white wine quality

colors = ['red', 'green', 'blue']

plt.bar(bar_labels, qualities, color=colors, edgecolor='black')
plt.title("Average Quality of Wine")
plt.xlabel("Quality by Wine Type")
plt.ylabel("Quality Score (0 - 10)");



```

```python
def quality_category(x):
    
    if x < 5:
        return "poor"
    elif x >= 5 and x <= 7:
        return "medium"
    else:
        return "high"
    
    
```
